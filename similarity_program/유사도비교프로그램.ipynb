```python
from sklearn.metrics.pairwise import cosine_similarity

def find_similar_rows_by_title(df, invention_title, num_similar_rows=10):
    """주어진 invention_title과 유사한 상위 N개의 행을 반환합니다."""

    # 주어진 invention_title을 가진 행의 인덱스 찾기
    target_row_index = df[df['invention_title'] == invention_title].index[0]

    # 선택한 행의 임베딩 벡터 가져오기
    selected_row_embedding = df.loc[target_row_index, 'kobert_embedding'].reshape(1, -1)

    # 모든 행에 대한 유사도 계산
    similarities = df['kobert_embedding'].apply(lambda emb: cosine_similarity(selected_row_embedding, emb.reshape(1, -1)).item())

    # 유사도에 따라 데이터 정렬 및 상위 N개 선택
    top_similar_indices = similarities.argsort()[::-1][1:(num_similar_rows+1)]
    top_similar_rows = df.loc[top_similar_indices]

    return top_similar_rows
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def find_similar_rows_by_title(df, invention_title, num_similar_rows=10):
    """주어진 invention_title과 유사한 상위 N개의 행을 반환합니다."""

    # 주어진 invention_title을 가진 행의 인덱스와 ipc_main 값 찾기
    target_row = df[df['invention_title'] == invention_title].iloc[0]
    target_row_index = target_row.name
    target_ipc_main = target_row['ipc_main']

    # 선택한 행의 임베딩 벡터 가져오기
    selected_row_embedding = target_row['kobert_embedding'].reshape(1, -1)

    # ipc_main 값이 일치하는 행만 필터링
    filtered_df = df[df['ipc_main'] == target_ipc_main].reset_index(drop=True)

    # 모든 필터링된 행에 대한 유사도 계산
    similarities = filtered_df['kobert_embedding'].apply(lambda emb: cosine_similarity(selected_row_embedding, emb.reshape(1, -1)).item())

    # 유사도에 따라 데이터 정렬 및 상위 N개 선택 (자기 자신 제외)
    similarities[target_row_index] = -1
    top_similar_indices = similarities.argsort()[::-1][:num_similar_rows]
    top_similar_rows = filtered_df.loc[top_similar_indices]

    return top_similar_rows
```

```python
# 랜덤으로 1000개의 invention_title 추출
sample_titles = df['invention_title'].sample(1000)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def find_similar_rows_by_title(df, invention_title, num_similar_rows=10):
    """주어진 invention_title과 유사한 상위 N개의 행을 반환합니다."""

    # 주어진 invention_title을 가진 행의 인덱스 찾기
    target_row_index = df[df['invention_title'] == invention_title].index[0]

    # 선택한 행의 임베딩 벡터 가져오기
    selected_row_embedding = df.loc[target_row_index, 'kobert_embedding'].reshape(1, -1)

    # 모든 행에 대한 유사도 계산
    similarities = df['kobert_embedding'].apply(lambda emb: cosine_similarity(selected_row_embedding, emb.reshape(1, -1)).item())

    # 유사도에 따라 데이터 정렬 및 상위 N개 선택
    # 유사도가 가장 높은 행(자기 자신)은 제외
    top_similar_indices = similarities.argsort()[::-1][1:(num_similar_rows+1)]
    
    # 유효한 인덱스만 선택
    valid_indices = [idx for idx in top_similar_indices if idx in df.index]
    top_similar_rows = df.loc[valid_indices]

    return top_similar_rows
```

```python
# 유사도 검사 및 라벨 일치 여부 확인
accuracy_count = 0

for title in sample_titles:
    similar_rows = find_similar_rows_by_title(df, title)
    input_label = df[df['invention_title'] == title]['ipc_subclass_num'].iloc[0]
    
    # 출력된 각 행의 라벨이 입력 라벨과 일치하는지 확인
    accuracy_count += np.sum(similar_rows['ipc_subclass_num'] == input_label)
    print(title)

# 테스트 정확도 계산
test_accuracy = accuracy_count / (1000 * 10)
```
