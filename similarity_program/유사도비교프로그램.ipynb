```python
import numpy as np

# 문자열을 배열로 변환하는 함수
def convert_embedding_str_to_array(embed_str):
    # 문자열에서 줄바꿈과 대괄호를 제거하고, 공백을 구분자로 사용하여 배열로 변환
    return np.fromstring(embed_str.replace('\n', '').replace('[', '').replace(']', ''), sep=' ')

# 모든 임베딩을 배열로 변환
# df_merge['kobert_embedding'] = df_merge['kobert_embedding'].apply(convert_embedding_str_to_array)
```

```python
df['kobert_embedding'] = df['kobert_embedding'].apply(convert_embedding_str_to_array)
```

```python
# CSV 파일 경로 지정
csv_file_path = './train_data_v1.3.csv'

# CSV 파일 불러오기
labeling_df = pd.read_csv(csv_file_path, sep=",", quotechar='"',encoding='utf-8', engine='python')

# 데이터프레임을 확인
labeling_df
```

```python
merged_df = pd.merge(df, labeling_df[['invention_title', 'ipc_subclass_num','ipc_main']], on='invention_title', how='left')
```

```python
df = merged_df
df.drop_duplicates(subset=['invention_title'],inplace=True)
df.reset_index(drop=True)
```

```python
def extract_part(text):
    try:
        # '-'를 기준으로 분리하고, 두 번째 부분을 '/'로 다시 분리
        parts = text.split('-')[0]
        # 필요한 부분 추출
        return parts
    except IndexError:
        # 형식이 맞지 않는 경우 None 반환
        return None

# DataFrame의 'text' 컬럼에 함수 적용
df['extracted_part'] = df['ipc_main'].apply(extract_part)
```

```python
def extract_three_letters(code):
    try:
        # '-'를 기준으로 분리하고, 두 번째 부분에서 앞의 세 글자 추출
        return code.split('-')[1][:3]
    except IndexError:
        # 형식이 맞지 않는 경우 None 반환
        return None
```

```python
# DataFrame의 'code' 컬럼에 함수 적용
df['ipc_main_detail'] = df['ipc_main'].apply(extract_three_letters)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def find_similar_rows_by_title(df, invention_title, num_similar_rows=10):
    """주어진 invention_title과 유사한 상위 N개의 행을 반환합니다."""

    # 주어진 invention_title을 가진 행의 인덱스 찾기
    target_row_index = df[df['invention_title'] == invention_title].index[0]

    # 선택한 행의 임베딩 벡터 가져오기
    selected_row_embedding = df.loc[target_row_index, 'kobert_embedding'].reshape(1, -1)

    # 모든 행에 대한 유사도 계산
    similarities = df['kobert_embedding'].apply(lambda emb: cosine_similarity(selected_row_embedding, emb.reshape(1, -1)).item())

    # 유사도에 따라 데이터 정렬 및 상위 N개 선택
    top_similar_indices = similarities.argsort()[::-1][1:(num_similar_rows+1)]
    top_similar_rows = df.loc[top_similar_indices]

    return top_similar_rows
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def find_similar_rows_by_title(df, invention_title, num_similar_rows=11):
    """주어진 invention_title과 유사한 상위 N개의 행을 반환합니다."""

    # 주어진 invention_title을 가진 행의 인덱스와 ipc_main 값 찾기
    target_row = df[df['invention_title'] == invention_title].iloc[0]
    target_row_index = target_row.name
    target_ipc_main = target_row['extracted_part'] #target_row['ipc_main']

    # 선택한 행의 임베딩 벡터 가져오기
    selected_row_embedding = target_row['kobert_embedding'].reshape(1, -1)

    # ipc_main 값이 일치하는 행만 필터링
    filtered_df = df[df['extracted_part'] == target_ipc_main].reset_index(drop=True) # df[df['ipc_main'] == target_ipc_main].reset_index(drop=True)

    # 모든 필터링된 행에 대한 유사도 계산
    similarities = filtered_df['kobert_embedding'].apply(lambda emb: cosine_similarity(selected_row_embedding, emb.reshape(1, -1)).item())

    # 유사도에 따라 데이터 정렬 및 상위 N개 선택 (자기 자신 제외)
    similarities[target_row_index] = -1
    top_similar_indices = similarities.argsort()[::-1][:num_similar_rows]
    top_similar_rows = filtered_df.loc[top_similar_indices]

    return top_similar_rows[1:]
```

```python
find_similar_rows_by_title(df, "카드 게임 서비스 제공 방법") # ipc_main 필터링o
```

```python
find_similar_rows_by_title(df, "시각 장애인용 원격 교육 지원 시스템") # ipc_main 필터링o
```

```python
find_similar_rows_by_title(df, "온라인 게임 아이템 거래 시스템 및 방법") # ipc_main 필터링o
```

```python
# 랜덤으로 1000개의 invention_title 추출
sample_titles = df['invention_title'].sample(100)
```

```python
# 유사도 검사 및 라벨 일치 여부 확인
accuracy_count = 0

for title in sample_titles:
    sum_detail=0
    # 유사도 검사 및 라벨 일치 여부 확인
    similar_rows = find_similar_rows_by_title(df, title)
    input_label = df[df['invention_title'] == title]['ipc_main_detail'].iloc[0]

    main_label = df[df['invention_title'] == title]['extracted_part'].iloc[0]
    filtered_df = df[(df['extracted_part'] == main_label) & (df['ipc_main_detail'] == input_label)]
    
    # 출력된 각 행의 라벨이 입력 라벨과 일치하는지 확인
    sum_detail += np.sum(similar_rows['ipc_main_detail'] == input_label)
    accuracy_count += sum_detail
    print(title,sum_detail,"/",len(filtered_df))

# 테스트 정확도 계산
test_accuracy = accuracy_count / 1000
```
